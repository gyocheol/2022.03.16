```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### ë¶“ê½ƒ ë°ì´í„°ì˜ GMM êµ°ì§‘í™”
- n_componentsëŠ” 3ìœ¼ë¡œ ì§€ì •
- êµ°ì§‘í™” ê²°ê³¼ëŠ” irisDFì˜ 'gmm_cluster' í•„ë“œë¡œ ì¶”ê°€ ì €ìž¥
- GMM êµ°ì§‘ ê²°ê³¼ì™€ target ê°’ê³¼ ë¹„êµ

```python
iris = load_iris()
feature_names = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width' ]

iris_df = pd.DataFrame(data=iris.data, columns=feature_names)
iris_df['target'] = iris.target
iris_df.head()
```

- target 1ë§Œ cluster 2ë¡œ 45ê°œ(90%), 1ë¡œ 5ê°œ(10%) ë§¤í•‘
- target 0ëŠ” cluster 0ìœ¼ë¡œ, target 2ëŠ” cluster 1ë¡œ ëª¨ë‘ ë§¤í•‘

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3, random_state=0)
# ì•„ëž˜ ë‘ê°œë¥¼ í•œ ì¤„ë¡œ ì“¸ë ¤ë©´ .fit_predict()í•˜ë©´ë¨.
gmm.fit(iris.data)
cluster_labels = gmm.predict(iris.data)
print('cluster labels : ', np.unique(cluster_labels, return_counts=True))
```

```python
iris_df['gmm_cluster'] = cluster_labels
iris_df.head()
```

```python
iris_df.groupby(['target'])['gmm_cluster'].value_counts()
```

### K-í‰ê·  êµ°ì§‘í™” ìˆ˜í–‰

```python
# iris_df['kmeans_cluster']
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, random_state=0)
kmeans.fit(iris.data)
kmeans_cluster_labels = kmeans.predict(iris.data)
```

```python
np.unique(kmeans_cluster_labels, return_counts=True)
```

```python
iris_df['kmeans_cluster']=kmeans_cluster_labels
iris_df.head()
```

```python
iris_df.groupby(['target'])['kmeans_cluster'].value_counts()
```

## GMMì™€ K-í‰ê· ì˜ ë¹„êµ

- K-í‰ê· ì€ ì›í˜•ì˜ ë²”ìœ„ì—ì„œ êµ°ì§‘í™”ë¥¼ ìˆ˜í–‰í•¨
- ë°ì´í„° ì„¸íŠ¸ê°€ ì›í˜•ì˜ ë²”ìœ„ë¥¼ ê°€ì§ˆìˆ˜ë¡ KMeans êµ°ì§‘í™” íš¨ìœ¨ì´ ë”ìš± ë†’ì•„ì§

### ì˜ˆ. make_blobs()ë¡œ  cluster_std=0.5ì¸ êµ°ì§‘ 3ê°œë¥¼ ìƒì„±

```python
### í´ëŸ¬ìŠ¤í„° ê²°ê³¼ë¥¼ ë‹´ì€ DataFrameê³¼ ì‚¬ì´í‚·ëŸ°ì˜ Cluster ê°ì²´ë“±ì„ 
### ì¸ìžë¡œ ë°›ì•„ í´ëŸ¬ìŠ¤í„°ë§ ê²°ê³¼ë¥¼ ì‹œê°í™”í•˜ëŠ” í•¨ìˆ˜  

def visualize_cluster_plot(clusterobj, dataframe, label_name, iscenter=True):
    if iscenter :
        centers = clusterobj.cluster_centers_
        
    unique_labels = np.unique(dataframe[label_name].values)
    markers=['o', 's', '^', 'x', '*']
    isNoise=False

    for label in unique_labels:
        label_cluster = dataframe[dataframe[label_name]==label]
        if label == -1:
            cluster_legend = 'Noise'
            isNoise=True
        else :
            cluster_legend = 'Cluster '+str(label)
        
        plt.scatter(x=label_cluster['ftr1'], y=label_cluster['ftr2'], s=70,\
                    edgecolor='k', marker=markers[label], label=cluster_legend)
        
        if iscenter:
            center_x_y = centers[label]
            plt.scatter(x=center_x_y[0], y=center_x_y[1], s=250, color='white',
                        alpha=0.9, edgecolor='k', marker=markers[label])
            plt.scatter(x=center_x_y[0], y=center_x_y[1], s=70, color='k',\
                        edgecolor='k', marker='$%d$' % label)
    if isNoise:
        legend_loc='upper center'
    else: legend_loc='upper right'
    
    plt.legend(loc=legend_loc)
    plt.show()
```

**ê¸¸ê²Œ ëŠ˜ì–´ë‚œ íƒ€ì›í˜•ì˜ ë°ì´í„° ì…‹ì„ ìƒì„±**

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=0.5, random_state=0)
```

```python
# ê¸¸ê²Œ ëŠ˜ì–´ë‚œ íƒ€ì›í˜• ë°ì´í„°ì…‹ ìƒì„±ì„ ìœ„í•œ ë³€í™˜
trans = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X_ansio = np.dot(X, trans)

cluster_df = pd.DataFrame(data=X_ansio, columns=['ftr1', 'ftr2'])
cluster_df['target'] = y
cluster_df.head()
```

```python
visualize_cluster_plot(None, cluster_df, 'target', iscenter=False)
```

### KMeansë¥¼ ì ìš©í•˜ì—¬ íƒ€ì›í˜• ë°ì´í„°ì…‹ì— ëŒ€í•œ êµ°ì§‘í™” ê²°ê³¼

```python
kmeans = KMeans(n_clusters=3, random_state=0)
cluster_labels = kmeans.fit_predict(X_ansio)
cluster_df[('Kmeans_cluster')]=cluster_labels
np.unique(cluster_labels, return_counts=True)
```

```python
visualize_cluster_plot(kmeans, cluster_df, 'Kmeans_cluster', iscenter=True)
```

### GMMì„ ì ìš©í•œ êµ°ì§‘í™”

```python
gmm = GaussianMixture(n_components=3, random_state=0)
gmm_cluster_labels = gmm.fit(X_ansio).predict(X_ansio)
cluster_df['gmm_cluster'] = gmm_cluster_labels
np.unique(gmm_cluster_labels, return_counts=True)
```

```python
visualize_cluster_plot(gmm, cluster_df, 'gmm_cluster', iscenter=False)
```

### K-í‰ê· ê³¼ GMM êµ°ì§‘í™” ê²°ê³¼ ë¹„êµ

```python
cluster_df.groupby(['target'])['Kmeans_cluster'].value_counts()
```
